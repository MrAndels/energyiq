```python
# ==========================================
# IMPORTACIÓN DE LIBRERÍAS
# ==========================================

# Manejo de datos
import pandas as pd

# Visualización
import matplotlib.pyplot as plt
```

```python
# ==========================================
# CARGAR MÉTRICAS
# ==========================================

# Linear Regression
lr = pd.read_csv(
    "../results/linear_regression_metrics.csv"
)

# Random Forest
rf = pd.read_csv(
    "../results/random_forest_metrics.csv"
)

# Neural Network
nn = pd.read_csv(
    "../results/neural_network_metrics.csv"
)

# Unir dataframes
comparison_df = pd.concat(
    
    [lr, rf, nn],
    
    ignore_index=True
)

# Mostrar resultados
comparison_df
```

```python
# ==========================================
# COMPARACIÓN R2
# ==========================================

# Crear figura
plt.figure(figsize=(8,5))

# Crear barras
plt.bar(
    
    # Modelos
    comparison_df["Modelo"],
    
    # Valores R2
    comparison_df["R2"]
)

# Título
plt.title(
    "Comparación R2 entre modelos"
)

# Eje X
plt.xlabel("Modelo")

# Eje Y
plt.ylabel("R2 Score")

# Mostrar gráfica
plt.show()
```

```python
# ==========================================
# COMPARACIÓN RMSE
# ==========================================

# Crear figura
plt.figure(figsize=(8,5))

# Crear barras
plt.bar(
    
    # Modelos
    comparison_df["Modelo"],
    
    # Valores RMSE
    comparison_df["RMSE"]
)

# Título
plt.title(
    "Comparación RMSE entre modelos"
)

# Eje X
plt.xlabel("Modelo")

# Eje Y
plt.ylabel("RMSE")

# Mostrar gráfica
plt.show()
```

```python
# ==========================================
# MEJOR MODELO
# ==========================================

# Obtener fila del mejor modelo
best_model = comparison_df.loc[
    
    # Buscar mayor R2
    comparison_df["R2"].idxmax()
]

# Mostrar resultado
print("Mejor modelo encontrado:\n")

print(best_model)
```
