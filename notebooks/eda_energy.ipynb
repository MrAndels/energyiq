```python
# Importar librerías para análisis de datos
import pandas as pd
import numpy as np
```

```python
# URL RAW del dataset en GitHub
url = "https://raw.githubusercontent.com/MrAndels/energyiq/refs/heads/main/data/raw/energy_dataset_reducido.csv"

# Cargar dataset desde GitHub
df = pd.read_csv(
    url,
    
    # Separador de columnas
    sep=";",
    
    # Evita problemas de memoria
    low_memory=False
)
```

```python
# ==========================================
# ESTRUCTURA GENERAL DEL DATASET
# ==========================================

# Mostrar primeras 5 filas
print("Primeras 5 filas:")
print(df.head())

# Mostrar dimensiones del dataset
print("\nDimensiones del dataset:")
print(df.shape)

# Mostrar nombres de columnas
print("\nColumnas del dataset:")
print(df.columns)

# Mostrar información general
print("\nInformación del dataset:")
print(df.info())

# Mostrar memoria utilizada
print("\nUso de memoria:")
print(df.memory_usage(deep=True))
```

```python
# ==========================================
# TIPOS DE VARIABLES
# ==========================================

# Mostrar tipos de datos de cada columna
print("Tipos de datos:")
print(df.dtypes)

# Contar cantidad de tipos de variables
print("\nCantidad de tipos de datos:")
print(df.dtypes.value_counts())

# Mostrar columnas tipo texto
print("\nColumnas tipo texto:")
print(df.select_dtypes(include="object").columns)

# Mostrar columnas numéricas
print("\nColumnas numéricas:")
print(df.select_dtypes(include=["float64", "int64"]).columns)
```

```python
# ==========================================
# VALORES NULOS
# ==========================================

# Contar valores nulos por columna
print("Valores nulos por columna:")
print(df.isnull().sum())

# Calcular porcentaje de valores nulos
print("\nPorcentaje de valores nulos:")

# Mostrar porcentaje por columna
print((df.isnull().sum() / len(df)) * 100)
```

```python
# ==========================================
# ESTADÍSTICAS Y DISTRIBUCIÓN DE VARIABLES
# ==========================================

# Mostrar estadísticas descriptivas
print("Resumen estadístico del dataset:")

# Analizar variables numéricas
print(df.describe(include="all"))
```

```python
# ==========================================
# DUPLICADOS E INCONSISTENCIAS
# ==========================================

# Contar filas duplicadas
duplicados = df.duplicated().sum()

# Mostrar cantidad de duplicados
print("Cantidad de filas duplicadas:")
print(duplicados)

# Mostrar primeras filas duplicadas
print("\nPrimeras filas duplicadas:")

# Filtrar registros duplicados
print(df[df.duplicated()].head())
```

```python
# ==========================================
# CORRELACIONES ENTRE VARIABLES
# ==========================================

# Lista de columnas numéricas
columnas_numericas = [
    "Global_active_power",
    "Global_reactive_power",
    "Voltage",
    "Global_intensity",
    "Sub_metering_1",
    "Sub_metering_2",
    "Sub_metering_3"
]

# Convertir columnas a tipo numérico
for columna in columnas_numericas:
    
    # Convertir texto a números
    df[columna] = pd.to_numeric(
        df[columna],
        
        # Convertir errores en NaN
        errors="coerce"
    )

# Calcular correlaciones
correlaciones = df[columnas_numericas].corr()

# Mostrar matriz de correlaciones
print("Correlaciones entre variables:")
print(correlaciones)
```
